```python
import spacy
spacy_md = spacy.load('en_core_web_md', disable=['parser', 'tagger', 'textcat'])
```

```python
import nltk
nltk.download('movie_reviews')
nltk.download('punkt')
```

```python
def ner_spacy(text):
    entities = []
    labels = []

    spacy_doc = spacy_md(text)
    
    for ent in spacy_doc.ents:
        entities.append(ent.text)
        labels.append(ent.label_)

    entities_labels = zip(entities, labels) 
    dataframe = pd.DataFrame({'entities': entities,'labels': labels})
    # dataframe = dataframe[dataframe['labels'].isin(['PERSON','ORG'])]
    dfy = dataframe.groupby('entities').count().sort_values('labels', ascending=False).head(10).reset_index()
    top_ents = dict(dfy.values)

    return top_ents
```
